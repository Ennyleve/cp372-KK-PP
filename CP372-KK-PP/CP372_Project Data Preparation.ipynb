# Part II : Data Preparation

## Part II.I : ตรวจสอบและแก้ไขค่าว่าง

```python
# ตรวจสอบข้อมูลว่างเปล่าภายใน Data

print(df.isnull().sum())
```

```python
# 1. เปลี่ยนค่า NaN ในคอลัมน์ ' College ' เป็น [ Unknown ]
df["College"] = df["College"].fillna("Unknown")

# - - - - - - - - - - - - - - -

# 2. เปลี่ยนค่า NaN ในคอลัมน์ ' Salary ' เป็น [ ค่ากลาง ( Median ) ]
df["Salary"] = df["Salary"].fillna(df["Salary"].median())

# - - - - - - - - - - - - - - -

# 3. เปลี่ยนค่า NaN ในคอลัมน์ 'Points', 'Rebounds', 'Assists' เป็น [ 0 ]
df["Points"] = df["Points"].fillna(0)
df["Rebounds"] = df["Rebounds"].fillna(0)
df["Assists"] = df["Assists"].fillna(0)
```

```python
print(df.isnull().sum())
```

```python
print(df.to_string())
```

## Part II.II : เพิ่ม Column ที่ต่อยอดได้

```python
# แปลงข้อมูล " ความสูง " จากเดิมที่มีหน่วยเป็น [ Foot-Inch ] ให้เป็น [ Meter ]

def feet_decimal_to_meters(height_float):
    try:
        meters = height_float * 0.3048  # 1 ฟุต = 0.3048 เมตร
        return round(meters, 2)
    except:
        return None

df[" Height ( meter ) "] = df["Height_i"].apply(feet_decimal_to_meters)
```

```python
print(df.to_string())
```

```python
# แปลงข้อมูล " น้ำหนัก " จากเดิมที่มีหน่วยเป็น [ Pound/lbs ] ให้เป็น [ Kg ]

def pounds_to_kg(weight):
    try:
        return round(weight * 0.453592, 1)
    except:
        return None

df[" Weight ( Kg ) "] = df["Weight"].apply(pounds_to_kg)
```

```python
print(df.to_string())
```

```python
# เพิ่ม Column แสดงค่า BMI ( ค่าดัชนีมวลกาย ) โดยนำข้อมูล [ ความสูง ] และ [ น้ำหนัก ] ที่มีการเตรียมข้อมูลไว้ก่อนหน้า ด้วยการแปลงหน่วยเอาไว้ .. เพื่อใช้ในการคำนวนอย่างถูกต้อง

def calculate_bmi(row):
    try:
        weight = row[" Weight ( Kg ) "]
        height = row[" Height ( meter ) "]
        if height > 0:
            return round(weight / (height ** 2), 1)
        else:
            return None
    except:
        return None

df["BMI"] = df.apply(calculate_bmi, axis=1)
```

```python
print(df.to_string())
```

```python
# จากค่า BMI ( ค่าดัชนีมวลกาย ) ที่ได้จากขั้นตอนที่แล้ว .. สามารถนำมาต่อยอดได้ด้วยการเพิ่มการแสดงความหมายของตัวเลขกำกับไว้ด้านหลังสุด ว่าค่าดัชนีมวลกายของนักกีฬาแต่ละคนมีความแตกต่างกันในช่วงใดบ้าง

def bmi_category(bmi):
    if bmi is None:
        return "Unknown"
    elif bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Overweight"
    elif bmi < 35:
        return "Obese I"
    elif bmi < 40:
        return "Obese II"
    else:
        return "Obese III"

df["BMI Category"] = df["BMI"].apply(bmi_category)
```

```python
print(df.to_string())
```
